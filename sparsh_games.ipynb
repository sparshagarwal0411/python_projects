```python
#3. MYSTERY WORD

import random

GREEN = "\033[92m"
YELLOW = "\033[93m"
RED = "\033[0;31m"
PURPLE = "\033[0;35m"
BLUE = "\033[0;34m"
BOLD = "\033[1m"
RESET = "\033[0m"

wordbank = [
    "about", "other", "which", "their", "there", "first", "would", "these", "click", "price", "state", "email",
    "world", "music", "after", "video", "where", "books", "links", "years", "order", "items", "group", "under",
    "games", "could", "great", "hotel", "store", "terms", "right", "local", "those", "using", "phone", "forum",
    "based", "black", "check", "index", "being", "women", "today", "south", "pages", "found", "house", "photo",
    "power", "while", "three", "total", "place", "think", "north", "posts", "media", "since", "guide", "board",
    "white", "small", "times", "sites", "level", "hours", "image", "title", "shall", "class", "still", "money",
    "every", "visit", "tools", "reply", "value", "press", "learn", "print", "stock", "point", "sales", "large",
    "table", "start", "model", "human", "movie", "march", "yahoo", "going", "study", "staff", "again", "april",
    "never", "users", "topic", "below", "zebra", "arise", "zeros"
]

user_name = input("What is your name?: ")
print(f"Welcome! to Mystery Word, {user_name}")
print(RED + "***RULES***" + RESET)
print("You have to guess a 5-letter word.")
print("You will get 5 attempts. Each attempt is important.")
print("After submitting a word you'll get hint in form of Result.")
print("If letter gets " + GREEN + "GREEN" + RESET + ", it means the letter is at the right spot.")
print("If letter gets " + YELLOW + "YELLOW" + RESET + ", it means the letter is in the word but not at the right spot.")
print("Type Yes if you want to continue. ")

y = input("Ready?: ").upper()

if y == "YES":
    w = random.choice(wordbank)
    word = w.upper()

    def wordle():
        guesses = []
        for x in range(1, 6):
            g = input(f"Enter your Guess No.{x}: ").upper()

            if g in guesses:
                print("⚠️ You already guessed that word. Attempt has been wasted.")
                continue
            guesses.append(g)

            if len(g) == 5:
                output = []

                for i in range(5):
                    if g[i] == word[i]:
                        output.append(GREEN + g[i] + RESET)
                    elif g[i] in word:
                        output.append(YELLOW + g[i] + RESET)
                    else:
                        output.append("-")

                result = "".join(output)
                print(f"Your word: {g}")
                print("Result:", result)

                if g == word:
                    print(f"🎉 Correct! You guessed the word {word} in your guess No.{x}.")
                    break
            else:
                print("The game ended because you did not enter a 5-letter word.")
                print(f"The word was: {word}")
                print("Thanks for playing.")
                print(PURPLE + "This game was made by" + RESET  + BLUE + BOLD + " BrainStormer " + RESET + PURPLE + "group. Found by " + RED + BOLD + "Sparsh" + RESET + PURPLE + " Make sure to check out other games on our website.")
                print("Rerun the code for new words. 🙃")
                break
        else:
            print("❌ You were not able to guess the word.")
            print(f"The word was: {word}")
            print("Hope you liked it! Thanks for playing.")
            print(PURPLE + "This game was made by" + RESET  + BLUE + BOLD + " BrainStormer " + RESET + PURPLE + "group. Found by " + RED + BOLD + "SPARSH" + RESET + PURPLE + " Make sure to check out other games on our website.")
            print("Rerun the code for new words. 🙃")

    wordle()
else:
    print("Type Yes please..")
```

```python
#4. FLAMES GAME

GREEN = "\033[92m"
YELLOW = "\033[93m"
RED = "\033[0;31m"
PURPLE = "\033[0;35m"
BLUE = "\033[0;34m"
BOLD = "\033[1m"
RESET = "\033[0m"
CYAN = "\033[0;36m"

def game():
  n1 = input("Enter name of 1st person: ").upper()
  n2 = input("Enter name of 2nd person: ").upper()

  l1 = len(n1)
  l2 = len(n2)

  if l1 >= l2:
    c = []

    for i in range(l1):
      if n1[i] in n2:
        c.append(n1[i])
        c1 = set(c)
        c2 = list(c1)

      else:
        continue

  else:
    c = []

    for i in range(l2):
      if n2[i] in n1:
        c.append(n2[i])
        c1 = set(c)
        c2 = list(c1)

      else:
        continue

  l = len(c2)

  lf = l1 + l2
  t = (lf - (2*l))

  flames_box = ["F", "L", "A", "M", "E", "S"]

  index = 0

  while len(flames_box) > 1:
    index = (index +t - 1) % len(flames_box)
    flames_box.pop(index)

  result = flames_box[0].upper()

  if result == "F":
    print(GREEN + "F" + RESET + RED + "LAMES" + RESET)
    print(CYAN + "FRIENDS" + RESET + "(FRIEND ZONED) 💔😢")

  elif result == "L":
    print(RED + "F" + RESET + GREEN + "L" + RESET + RED + "AMES" + RESET)
    print(CYAN + "LOVERS" + RESET + "(Perfect Match) ❤️")

  elif result == "A":
    print(RED + "FL" + RESET + GREEN + "A" + RESET + RED + "MES" + RESET)
    print(CYAN + "AFFECTIONATE" + RESET + "(Someone's in crush phase). ❤️")

  elif result == "M":
    print(RED + "FLA" + RESET + GREEN + "M" + RESET + RED + "ES" + RESET)
    print(CYAN + "MARRIAGE" + RESET + "(Get ready to get married) 🤵🏻❤️👰🏻")

  elif result == "E":
    print(RED + "FLAM" + RESET + GREEN + "E" + RESET + RED + "S" + RESET)
    print(CYAN + "ENEMIES" + RESET + "(Fighting against each other till death). 💔👿")

  else:
    print(RED + "FLAME" + RESET + GREEN + "S" + RESET)
    print(CYAN + "Siblings" + RESET + "(Brother - Sister) 💔😞")

  print(f'Thanks for playing {n1} and {n2}')
  print("Hope you liked it! Thanks for playing.")
  print(PURPLE + "This game was made by" + RESET  + BLUE + BOLD + " BrainStormer " + RESET + PURPLE + "group. Found by " + RED + BOLD + "SPARSH" + RESET + PURPLE + " Make sure to check out other games on our website.")


game()
```

```python
 #5. MATRIX MAKER
import numpy as np

def matrix():

  rows = int(input("Enter initial no of rows: "))
  columns = int(input("Enter initial no of columns: "))

  matrix = []
  for i in range(rows):
    row = []
    for j in range(columns):
      value = int(input(f"Enter element ({i+1},{j+1})"))
      row.append(value)
    matrix.append(row)

  print("Initial Matrix:")
  initial_arr = np.array(matrix)
  print(initial_arr)
```
